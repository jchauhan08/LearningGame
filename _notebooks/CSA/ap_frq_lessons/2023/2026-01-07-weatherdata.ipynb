import java.util.ArrayList;
import java.util.Arrays;
class Main
{
    private ArrayList<Double> temperatures;
    public Main(ArrayList<Double> temps)
    {
        this.temperatures = temps;
    }
    public void cleanData(double lower, double upper)
    {
        int i = 0;
        while (i < temperatures.size())
        {
            if (temperatures.get(i) < lower || temperatures.get(i) > upper)
            {
                temperatures.remove(i);
            }
            else
            {
                i++;
            }
        }
    }
    public void printTemperatures()
    {
        System.out.println(temperatures);
    }
    public static void main(String[] args)
    {
        // Test case 1
        ArrayList<Double> temps1 = new ArrayList<>(Arrays.asList(99.1, 142.0, 85.0, 85.1, 84.6, 94.3, 124.9, 98.0, 101.0, 102.5));
        Main weather1 = new Main(temps1);
        System.out.println("Original temperatures: ");
        weather1.printTemperatures();
        weather1.cleanData(80.0, 100.0);

        weather1.printTemperatures();
        System.out.println("Expected: [99.1, 85.0, 85.1, 84.6, 94.3, 98.0]");

{% capture solution_part_b %}
import java.util.ArrayList;
import java.util.Arrays;
class Main
{
    private ArrayList<Double> temperatures;
    public Main(ArrayList<Double> temps)
    {
        this.temperatures = temps;
    }
    public int longestHeatWave(double threshold)
    {
        int maxLength = 0;
        int currentLength = 0;
        for (double temp : temperatures)
        {
            if (temp > threshold)
            {
                currentLength++;
                if (currentLength > maxLength)
                {
                    maxLength = currentLength;
                }
            }
            else
            {
                currentLength = 0;
            }
        }
        return maxLength;
    }
    public static void main(String[] args)
    {
        ArrayList<Double> temps = new ArrayList<>(Arrays.asList(100.5, 98.5, 102.0, 103.9, 87.5, 105.2, 90.3, 104.6, 96.0));
        Main weather = new Main(temps);
        System.out.println("Temperatures: " + temps);
        System.out.println();
        // Test case 1
        int result1 = weather.longestHeatWave(100.5);
        System.out.println("longestHeatWave(100.5)");
        System.out.println("Expected: 3");
        System.out.println("Result: " + result1);
        System.out.println();
        // Test case 2
        int result2 = weather.longestHeatWave(95.0);
        System.out.println("longestHeatWave(95.0)");
        System.out.println("Expected: 4");
        System.out.println("Result: " + result2);
    }
}
